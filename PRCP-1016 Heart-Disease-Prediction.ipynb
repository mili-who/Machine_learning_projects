## ***Brief explanation of the features or individual columns :***

There are 15 columns in the dataset, where the 'patient_id' column is a unique and random identifier. The remaining 14 features are described in the section below.

* **Feature or Attribute columns**

1. slope_of_peak_exercise_st_segment : the slope of the peak exercise ST(Segment and T-wave) segment, an electrocardiography read out indicating quality of blood flow to the heart.
2. thal : results of thallium stress test measuring blood flow to the heart, with possible values -
    - 0 : normal  
    - 1 : fixed_defect and
    - 2 : reversible_defect
3. resting_blood_pressure : indicates the blood pressure level when a person is at rest.
4. chest_pain_type : refers to the classification of chest pain into four categories:
   - Typical angina - Discomfort related to the heart, triggered by exertion and relieved by rest or medication.
   - Atypical angina - Chest pain with some features of typical angina but not fitting all criteria.
   - Non-anginal pain - Chest pain that is not related to the heart and may be caused by other factors.
   - Asymptomatic - No chest pain reported or observed.
5. num_major_vessels : number of major vessels (0-3) colored by flourosopy. It helps assess the severity of heart disease and blockages in coronary arteries.The values 0-3 represent the number of major vessels colored during the fluoroscopy examination:
   - 0 : No major vessels colored, indicating no significant blockages.
   - 1 : One major vessel colored, suggesting a mild blockage in a single vessel.
   - 2 : Two major vessels colored, indicating moderate blockages in two vessels.
   - 3 : Three major vessels colored, suggesting severe blockages in multiple vessels.
6. fasting_blood_sugar_gt_120_mg_per_dl : fasting blood sugar > 120 mg/dl.
7. resting_ekg_results : resting electrocardiographic results (values 0,1,2)
   - 0: Normal - The resting electrocardiogram (ECG) is considered normal, showing no significant abnormalities.
   - 1: Abnormal - The resting ECG indicates minor abnormalities or changes that might require further evaluation.
   - 2: Suspicious - The resting ECG shows significant abnormalities or changes that raise suspicion of heart-related issues, warranting closer examination and medical attention.
8. serum_cholesterol_mg_per_dl : serum cholestoral in mg/dl.
9. oldpeak_eq_st_depression : oldpeak = ST depression induced by exercise relative to rest, a measure of abnormality in electrocardiograms.
10. sex : 0: female, 1: male
11. age : age in years
12. max_heart_rate_achieved : maximum heart rate achieved (beats per minute)
13. exercise_induced_angina : exercise-induced chest pain (0: False, 1: True)

* **Target column**

14. heart_disease_present : 0: heart-disease not present, 1: heart-disease present

# MODEL EVALUATION

```python
models = {
    "Logistic Regression":LogisticRegression(),
    "SVC":SVC(),
    "K Nearest Neighbor" : KNeighborsClassifier(),
    "Random Forest Classifier" : RandomForestClassifier(),
    'Decision Tree Classifier': DecisionTreeClassifier(),
    "XGBClassifier":XGBClassifier(),
    "Ada Boost Classifier" : AdaBoostClassifier()
}

# Create a function to fit and score models
def fit_and_score(models, X_train, X_test, y_train, y_test):

    np.random.seed(42)

    model_scores = {}

    for name, model in models.items():

        model.fit(X_train, y_train)

        model_scores[name] = model.score(X_test, y_test)
    return model_scores
```

```python
model_scores = fit_and_score(models = models,
                             X_train = X_train,
                             X_test = X_test,
                             y_train = y_train,
                             y_test = y_test)
```

```python
model_scores
```

```python
plt.figure(figsize = (4,4))
score_graph = pd.DataFrame(model_scores, index=["accuracy"])
score_graph.T.plot.bar()
```

**Three models aree giving good performance and overall cross validation score. However, the first model Logistic Regression might not capture the non-linear relationships between the input and output features, as it assumes a linear relationship between them.**
